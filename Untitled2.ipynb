Q. 2) In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.



```python
data_com_mais_publicacao = df["Release_Date"].value_counts()[0:10]
data_com_mais_publicacao
```

```python
top_10_years = df['Release_Date'].value_counts().nlargest(10).index
df_top_10 = df[df['Release_Date'].isin(top_10_years)]
plt.figure(figsize=(15, 7))
plt.xticks(rotation=45)
sns.countplot(data=df_top_10, x='Release_Date', hue='Category', order=top_10_years)


```

Q. 5) Show Top 10 Directors, who gave the highest number of TV Shows & Movies to Netflix ?



```python
top_director = df["Director"].value_counts().nlargest(10).index
top_df = df[df["Director"].isin(top_director)]
plt.figure(figsize=(15,7))
plt.xticks(rotation=45)
sns.countplot(data = top_df,x = "Director",order=top_director,hue="Category")


```

Q. 6) Show all the Records, where "Category is Movie and Type is Comedies" or "Country is United Kingdom".



```python
df.loc[(df["Category"] == "Movie") & (df["Type"].str.contains("Comedies"))].count()
```

```python
df.loc[(df["Category"] == "Movie") & (df["Type"].str.contains("Comedies")) | (df["Country"] == "United Kingdom")].count()
```
